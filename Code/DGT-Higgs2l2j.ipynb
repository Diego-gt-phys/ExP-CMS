# **Search for Higgs Bosons in the H → ZZ → 2l2q Channel**

## **Project Overview**
This project focuses on searching for a Higgs boson in the mass range of **200-900 GeV** using CMS data and Monte Carlo simulations.

## **Objectives**

### **1. Monte Carlo Simulation Study**
*   **Selection Strategy:** Design a selection strategy based on Monte Carlo simulations to suppress background processes relative to the Higgs signal.
*   **Kinematic Distributions:** Study kinematic distributions to effectively discriminate between signal and background events.

### **2. CMS Data Analysis**
*   **Data Selection:** Analyze real CMS data samples (containing a simulated signal injection) by applying the selection criteria derived from the simulation studies.
*   **Background Comparison:** Compare observed data with background predictions to search for any excess attributable to a signal.
*   **Signal Characterization:** Fit the data to a combination of signal and background to determine key properties of the found Higgs boson signal, including:
    *   Mass and Width
    *   Number of Events
    *   Statistical Significance
    *   Production Cross-section

#### A.1) Study the Kinematic Variables


```python
def Compare_Variables(var, df_name_list, lum_df_list, label_list, x_range, bins, xl,dens=True):
    cols = ["r", "b", "g", "m"]
    bin_edges = np.linspace(x_range[0], x_range[1], bins)

    fig, ax = plt.subplots(figsize=(13, 8))

    for i in range(len(df_name_list)):
        weight = np.full(len(df_name_list[i]), lumi_data / lum_df_list[i])
        ax.hist(df_name_list[i][var], bins=bin_edges, weights=weight,
                histtype='step', color=cols[i], label=label_list[i], linewidth=1.5, density=dens)
    hep.cms.label("Preliminary", data=False, lumi=lumi_data, com=8, ax=ax)

    ax.set_xlabel(xl)
    if dens:
        ax.set_ylabel('Density')
    else:
        ax.set_ylabel('Events')
    ax.legend()
    plt.show()
```

##### Establish Kinematic cuts

```python
# Make a copy for safe keping
ZJets_cut = ZJets.copy()
Higgs350_cut = Higgs350.copy()
Higgs400_cut = Higgs400.copy()
Higgs500_cut = Higgs500.copy()

# Cut the simulation data using ptl1>=70GeV
for df in [ZJets_cut, Higgs350_cut, Higgs400_cut, Higgs500_cut]:
    df.drop(df[df['ptl1'] < 70].index, inplace=True)

# Compare new sizes
len(ZJets_cut)/len(ZJets), len(Higgs350_cut)/len(Higgs350), len(Higgs400_cut)/len(Higgs400), len(Higgs500_cut)/len(Higgs500)
```

```python
df_list = [ZJets, Higgs350, Higgs400, Higgs500]
df_cut_list = [ZJets_cut, Higgs350_cut, Higgs400_cut, Higgs500_cut]
lum_df_list = [lumi_zjets, lumi_higgs350, lumi_higgs400, lumi_higgs500]
label_list = ["Z+jets", "H ($m_H=350$ GeV)", "H ($m_H=400$ GeV)", "H ($m_H=500$ GeV)"]
var = "ptl1"
cols = ["r", "b", "g", "m"]

bin_edges  = np.linspace(0, 400, 50)
fig, ax = plt.subplots(figsize=(13, 8))

for i in range(len(df_list)):
    weight = np.full(len(df_list[i]), lumi_data / lum_df_list[i])
    weight_cut = np.full(len(df_cut_list[i]), lumi_data / lum_df_list[i])

    ax.hist(df_list[i][var], bins=bin_edges, weights=weight,
            histtype='step', color=cols[i], label=label_list[i],
            linestyle="dashed", linewidth=1.5, density=True)
    ax.hist(df_cut_list[i][var], bins=bin_edges, weights=weight_cut,
            histtype='step', color=cols[i], label=label_list[i],
            linestyle="solid", linewidth=1.5, density=True)

handles_list = [
    mpatches.Patch(facecolor="none", edgecolor="k", label=r"$p_T \geq 70$ GeV", linewidth=1.5, linestyle="solid"),
    mpatches.Patch(facecolor="none", edgecolor="k", label="No Cuts", linewidth=1.5, linestyle="dashed"),
    mpatches.Patch(facecolor="none", edgecolor=cols[0], label="Z+jets", linewidth=1.5),
    mpatches.Patch(facecolor="none", edgecolor=cols[1], label=r"H ($m_H=350$ GeV)", linewidth=1.5),
    mpatches.Patch(facecolor="none", edgecolor=cols[2], label=r"H ($m_H=400$ GeV)", linewidth=1.5),
    mpatches.Patch(facecolor="none", edgecolor=cols[3], label=r"H ($m_H=500$ GeV)", linewidth=1.5),
]

ax.legend(
    handles=handles_list,
    loc="upper right",
    ncol=2,
    frameon=False,
    fancybox=False,          # CMS uses sharp corners, not rounded
    edgecolor="black",
    framealpha=1,            # fully opaque — more common in CMS than semi-transparent
    labelspacing=0.3,
    handletextpad=0.5,
    handlelength=1.5,
    columnspacing=1.0,
)

hep.cms.label("Preliminary", data=False, lumi=lumi_data, com=8, ax=ax)

ax.set_ylabel('Density')
ax.set_xlabel("Leading Lepton $p_T$ (GeV)")
plt.show()
```
